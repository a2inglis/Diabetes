```python
# ***** converting drugs into drug classes *****
# using a function and a dictionary

# MEGLITINIDES

# Define the dictionary map for the drug
# Define a function to map string values to numeric values
def map_to_numeric(value):
    mapping = {
        'No': 0,
        'Steady': 1,
        'Down': 2,
        'Up': 3
    }
    return mapping.get(value, -1)  # Default to -1 for unknown values
    
def map_to_drug_class(row):
    values = [row['repaglinide'], row['nateglinide']]
    
    # Convert string values to numeric values 
    numeric_values = [map_to_numeric(value) for value in values]
    
    # Find the highest value and map it to the drug class
    highest_value = max(numeric_values)
    
    # Convert numbers back to labels No, Up, Down, Steady
    # Define the reverse mapping for drug class
    reverse_mapping = {
        0:'No',
        1:'Steady',
        2:'Down',
        3:'Up'
    }
    
    # Map the highest numeric value to the drug class
    drug_class = reverse_mapping.get(highest_value, 'Unknown')
        
    return drug_class

# Apply the function to create the Meglitinides column
data["Meglitinides"] = data.apply(map_to_drug_class, axis = 1)

print(data)
```
