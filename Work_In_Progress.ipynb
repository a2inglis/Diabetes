## Split X and Y Datasets

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.naive_bayes import MultinomialNB
```

```python
# Read in the dataset passed from the last module

X = pd.read_csv('Mod_3_X_data_to_pass.csv')

Y = pd.read_csv('Mod_3_Y_data_to_pass.csv')

display(X.head())
display(Y.head())
```

```python
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 403)
```

## Decision Tree Classifier

```python
# Turn on the decision tree classifier tool
Tree_classifier = DecisionTreeClassifier()

# Fit the classifier on the training data
Tree_classifier.fit(X_train, Y_train)

# Make predictions on the test data
Y_Tree_predict = Tree_classifier.predict(X_test)

# Evaluate the model
Tree_accuracy = accuracy_score(Y_test, Y_Tree_predict)
Tree_report = classification_report(Y_test, Y_Tree_predict)

# Print results
print(f'Accuracy of tree classifier: {Tree_accuracy}')
print(Tree_report)
```

## Random Forest Classifier

```python
# Turn on the Random Forest Tool
rf_classifier = RandomForestClassifier(n_estimators = 100, random_state = 403)

# Fit the classifier to the training data
rf_classifier.fit(X_train, Y_train)

# Make predictions on the test data
Y_RFpredict = rf_classifier.predict(X_test)

# Evaluate the model
RF_Accuracy = accuracy_score(Y_test, Y_RFpredict)
RF_report =  classification_report(Y_test, Y_RFpredict)

# Print results
print(f'Accuracy of random forest classifier: {RF_Accuracy}')
print(RF_report)
```
